# Sorting Data

```python
def sort_pair(df, person):

    data = df.to_numpy()
    row_index = 0
    preferred_rankings = {}

    for row in data:
    
        column_index = 0
        preferred_rankings_list = [] # for storing the temp list with the id number and 

        for life_expectancy in row:
            preferred_rankings_list.append((column_index+1, life_expectancy)) # id number is the column index + 1  
            column_index += 1

        # sorts the 2d array by the second value in each item of the array in descending order
        # https://stackoverflow.com/a/20099713
        preferred_rankings_list = sorted(preferred_rankings_list, key=lambda x: x[1], reverse=True)

        preferred_rankings[f"{person} {row_index+1}"] = preferred_rankings_list

        row_index += 1
    return preferred_rankings

def unsort_pair(df, person):

    data = df.to_numpy()
    row_index = 0
    preferred_rankings = {}

    for row in data:
    
        column_index = 0
        preferred_rankings_list = [] # for storing the temp list with the id number and 

        for life_expectancy in row:
            preferred_rankings_list.append((column_index+1, life_expectancy)) # id number is the column index + 1  
            column_index += 1


        preferred_rankings[row_index+1] = preferred_rankings_list

        row_index += 1
    return preferred_rankings
```

# Pairing Algorithm
Now that we have each patient's and each donors preferred pairs we can begin the pairing algorithm similiar to the one on wiki:

    Initialize all p  P and d  D to free
    while  free patient p who still has a Donor d to pair to do
        d := first Donor on p's list to whom p has not yet paired
        if d is free then
            (p, d) become paired
        else some pair (p', d) already exists
            if d prefers p to p' then
                p' becomes free
                (p, d) become paired 
            else
                (p', d) remain paired
            end if
        end if
    repeat

```python
# Code in this cell heavily inspired by @Author: Ryan Schachte
# https://github.com/Schachte/stable-matching-algorithm

# Also heavily inspired by the Gale-Shapley algorithm 
# https://en.wikipedia.org/wiki/Gale%E2%80%93Shapley_algorithm
# and https://gist.github.com/joyrexus/9967709

import collections

# #Keep track of the people that "may" end up together
tentative_engagements = []

# #Patients who still need to propose and get accepted successfully
free_patients = []

def init_free_patients():
    '''Initialize the arrays of donors and patients to represent 
        that they're all initially free and not paired'''
    for patient in preferred_rankings_patients:
        free_patients.append(patient)
    

def calculateLifeExpect(patient, donor):
    listOfPatientsForDonor = preferred_rankings_patients.get(patient)
    
    # if donor on waiting list retutn a life expectancy of 10 years
    if donor == 'w':
        return 10
    
    lifeExpectancy = [y for (x, y) in listOfPatientsForDonor if x == donor][0]

    return lifeExpectancy

def begin_matching(patient):
    '''Find the first free donor available to a patient at
        any given time'''


    for donor_life in preferred_rankings_patients[patient]:
        
        #Boolean for whether donor is taken or not
        taken_pair = [pair for pair in tentative_engagements if donor_life[0] in pair]
        
        if (len(taken_pair) == 0):
            
            tentative_engagements.append([patient, donor_life[0]])
            free_patients.remove(patient)
            break

        # changed to less than
        elif (len(taken_pair) > 0):

            current_p = taken_pair[0][0]

            res1 = calculateLifeExpect(current_p,donor_life[0])
#             print("res1", res1)

            potential_p = patient

            res2 = calculateLifeExpect(patient, donor_life[0])
#             print(res2, "res2")
            
            if (res1 < res2):
                #print('choosing new patient, removing old one ')
                #The new patient is paired
                free_patients.remove(patient)

                #The old patient is now single
                free_patients.append(taken_pair[0][0])
                
                #Update the patient of the donor (tentatively)
                taken_pair[0][0] = patient
                break

def stable_matching():
    '''Matching algorithm until stable match terminates'''
    while (len(free_patients) > 0):
        for patient in free_patients:
            begin_matching(patient)



init_free_patients()
stable_matching()

pd.DataFrame(tentative_engagements).head()
```

# Checking Results Functions



Checking for Duplicates

Function for checking Duplicates

* Returns False if no Duplicates found
* Returns (True, First Donor/Patient Duplicate) if Duplicates found

```python
def check_duplicates(array):
    
    # check patients
    testSort = sorted(array, key= lambda x: int(x[0].split(" ")[1]), reverse=True)
    for i in range(0,len(testSort)-1):
        if (int(testSort[i][0].split(" ")[1]) == int(testSort[i+1][0].split(" ")[1])):
            return ("Duplicate Patient", testSort[i])
    
    # check donors
    testSort = sorted(array, key= lambda x: int(x[1]), reverse=True)
    for i in range(0,len(testSort)-1):
        if (testSort[i][1] == testSort[i+1][1]):
            return (True, i, testSort[i])
    
    # if no duplicates found
    return False
```

Adding to waiting list

Function for getting final array by adding patients with life expectancy below 10 to the waiting list

* if life expectancy below 10:
    * add to waiting list
    * change life expectancy to 10 years 

```python
import copy

def get_final_array(array):

    added_waiting_list = copy.deepcopy(array)
    waiting_free_donors = []
    life_expectancy = 0

    for patientDonorPair in added_waiting_list:
        # print(patientDonorPair, added_waiting_list.index(patientDonorPair))
        life_expectancy = calculateLifeExpect(patientDonorPair[0], patientDonorPair[1])
        
        if(life_expectancy < 10):
            waiting_free_donors.append(patientDonorPair[0])
            waiting_free_donors.append(patientDonorPair[1])
            patientDonorPair[1] = 'w'
            life_expectancy = 10
    
        if(len(patientDonorPair) == 3):
            patientDonorPair[2] = life_expectancy
        else:
            patientDonorPair.append(life_expectancy)

    return sorted(added_waiting_list, key=lambda x: x[2], reverse=True)
```

# In The Other File There Is A Top Trading Cycles Algorithm

- Step 1  All kidneys are available, and all agents are active. Each patient ti points either to his most preferred kidney or to the wait-list option, w. Each remaining kidney ki points to its paired patient, ti.

- Step 2  At this point, by Lemma 1, there is either a cycle or a w-chain, or both.
  - If there is not a cycle, proceed to step 3.
 - If there is a cycle, carry out the corresponding exchanges and remove all
patients and kidneys involved from the mechanism.
    - Once this has been done, have all remaining patients point to their
top choice among the remaining kidneys. Locate any new cycles,
remove them, and repeat until there are no more cycles.

- Step 3  At this point, all remaining pairs are the tails of w-chains.
    -  Select one of the chains according to whichever chain selection rule you choose. This assignment is final for those in the selected w-chain.
        - Depending upon which chain selection rule you use, the w-chain is either removed from the mechanism, or patients and kidneys remain in the mechanism, but are passive instead of active (cannot accept new assignments)
        
        
- Step 4  After the w-chain is selected, new cycles may form, so you must repeat steps 2 and 3 until there are no patients or kidneys are left.

**By the end of this procedure, every patient with a living donor has either been assigned a living kidney for transplant or a high priority position on the waiting list.**
