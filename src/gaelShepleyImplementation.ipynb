# Sorting Function:

```python
def sort_pair(df, person):

  data = df.to_numpy()
  row_index = 0
  preferred_rankings = {}

  for row in data:
    
    column_index = 0
    preferred_rankings_list = [] # for storing the temp list with the id number and 
    
    for life_expectancy in row:
        preferred_rankings_list.append((column_index+1, life_expectancy)) # id number is the column index + 1  
        column_index += 1
        
    # sorts the 2d array by the second value in each item of the array in descending order
    # https://stackoverflow.com/a/20099713
    preferred_rankings_list = sorted(preferred_rankings_list, key=lambda x: x[1], reverse=True)

    preferred_rankings[f"{person} {row_index+1}"] = preferred_rankings_list
    
    row_index += 1

  return preferred_rankings
```

```python
# Code in this cell heavily inspired by @Author: Ryan Schachte
# https://github.com/Schachte/stable-matching-algorithm
#Also heavily inspired by the Gale-Shapley algorithm 
#https://en.wikipedia.org/wiki/Gale%E2%80%93Shapley_algorithm
#and https://gist.github.com/joyrexus/9967709

import collections

# #Keep track of the people that "may" end up together
tentative_engagements = []

# #Patients who still need to propose and get accepted successfully
free_patients = []

def init_free_patients():
    '''Initialize the arrays of donors and patients to represent 
        that they're all initially free and not engaged'''
    for patient in preferred_rankings_patients:
        free_patients.append(patient)
    

def calculateLifeExpect(patient,K ):
    listOfPatientsForDonor = preferred_rankings_patients.get(patient)
    # k is current donor 
    res = [y for (x, y) in listOfPatientsForDonor if x == K] 
    return res 

def begin_matching(patient):
    '''Find the first free donor available to a patient at
        any given time'''


    for donor_life in preferred_rankings_patients[patient]:
        
        #Boolean for whether donor is taken or not
        taken_pair = [pair for pair in tentative_engagements if donor_life[0] in pair]
        
        if (len(taken_pair) == 0):
            
            tentative_engagements.append([patient, donor_life[0]])
            free_patients.remove(patient)
            break

        elif (len(taken_pair) > 0):

            #print key and value pair 
            current_p = taken_pair[0][0]

            res1 = calculateLifeExpect(current_p,donor_life[0])
#             print("res1", res1)

            potential_p = patient

            res2 = calculateLifeExpect(patient, donor_life[0])
#             print(res2, "res2")
            
            
            if (res1 < res2):
                #print('choosing new patient, removing old one ')
                #The new patient is engaged
                free_patients.remove(patient)

                #The old patient is now single
                free_patients.append(taken_pair[0][0])
                
#                 print(taken_pair)
                #Update the patient of the donor (tentatively)
                taken_pair[0][0] = patient
#                 print(taken_pair)
                break

def stable_matching():
    '''Matching algorithm until stable match terminates'''
    while (len(free_patients) > 0):
        for patient in free_patients:
            begin_matching(patient)



init_free_patients()
stable_matching()

pd.DataFrame(tentative_engagements)


```
