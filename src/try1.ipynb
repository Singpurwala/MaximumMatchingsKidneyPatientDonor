```python
#trim data to only table 

#get info from the first sheet (here we only have one sheet)
df = pd.read_excel(xls,sheet_names[0])
df2 = df.copy(deep=True)
df = df.drop(columns=['Pair', 'Patient Bloodtype','Donor Bloodtype','Compatibility','Unnamed: 4'])


new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header

df.head()
```

```python
df2.drop(df2.columns[3:305], axis=1, inplace=True)
df2.head()
```

```python
sorted_donor_data = []

data = df.to_numpy()

for row in data:
  
  column_index = 0
  life_donor_list = [] # for storing the temp list with the donor number and 
  
  for column in row:
    life_donor_list.append([column_index+1, column])
    column_index += 1
  
  life_donor_list = sorted(life_donor_list, key=lambda x: x[1], reverse=True) # sorts the 2d array by the second value in each item of the array in descending order

  sorted_donor_data.append(life_donor_list)

  # print(sorted_donor_data)
  # break

sorted_donor_dataframe = pd.DataFrame(sorted_donor_data)  
```
